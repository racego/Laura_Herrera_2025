```python
# GRAFICADOR DE INDICADORES AMBIENTALES Y SST - COLEGIO LAURA HERRERA
import matplotlib.pyplot as plt
import numpy as np

# Configuración inicial
plt.style.use('seaborn-v0_8')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# DATOS AMBIENTALES
categorias_amb = ['Reducción Residuos', 'Ahorro Energético', 'Incremento Áreas Verdes']
meta_amb = [30, 15, 20]  # Porcentajes meta
actual_amb = [12, 8, 5]   # Porcentajes actuales (simulados)
colores_amb = ['#2E8B57', '#FFA500', '#32CD32']

# Gráfico ambiental - Barras comparativas
x_pos = np.arange(len(categorias_amb))
ancho = 0.35

barras_meta = ax1.bar(x_pos - ancho/2, meta_amb, ancho, label='Meta 12 meses',
                      color=colores_amb, alpha=0.7, edgecolor='black')
barras_actual = ax1.bar(x_pos + ancho/2, actual_amb, ancho, label='Actual',
                        color=colores_amb, alpha=0.4, edgecolor='black')

ax1.set_title('INDICADORES AMBIENTALES\nColegio Laura Herrera - Bosa',
              fontsize=14, fontweight='bold', pad=20)
ax1.set_xlabel('Categorías')
ax1.set_ylabel('Porcentaje (%)')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(categorias_amb, rotation=15)
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Añadir valores en las barras
for i, (meta, actual) in enumerate(zip(meta_amb, actual_amb)):
    ax1.text(i - ancho/2, meta + 1, f'{meta}%', ha='center', va='bottom', fontweight='bold')
    ax1.text(i + ancho/2, actual + 1, f'{actual}%', ha='center', va='bottom', fontweight='bold')

# DATOS SST (Seguridad y Salud en el Trabajo)
categorias_sst = ['Señalización\nImplementada', 'Comunidad\nCapacitada', 'Accidentes\nEvitados']
valores_sst = [100, 80, 95]  # Porcentajes
colores_sst = ['#1E90FF', '#FF6347', '#9370DB']

# Gráfico SST - Barras horizontales
y_pos = np.arange(len(categorias_sst))

barras_sst = ax2.barh(y_pos, valores_sst, color=colores_sst, alpha=0.7, edgecolor='black')
ax2.set_title('INDICADORES SST (Seguridad y Salud)\nColegio Laura Herrera - Bosa',
              fontsize=14, fontweight='bold', pad=20)
ax2.set_xlabel('Porcentaje (%)')
ax2.set_yticks(y_pos)
ax2.set_yticklabels(categorias_sst)
ax2.set_xlim(0, 110)
ax2.grid(axis='x', alpha=0.3)

# Añadir valores en las barras horizontales
for i, valor in enumerate(valores_sst):
    ax2.text(valor + 2, i, f'{valor}%', va='center', fontweight='bold', fontsize=12)

# Línea de meta al 100%
ax2.axvline(x=100, color='red', linestyle='--', alpha=0.7, label='Meta 100%')
ax2.legend()

# AJUSTES FINALES
plt.tight_layout()
plt.suptitle('SISTEMA DE SEGUIMIENTO - OPORTUNIDADES DE MEJORA AMBIENTAL Y SST',
             fontsize=16, fontweight='bold', y=1.02)
plt.show()

# TABLA RESUMEN DE INDICADORES
print("="*70)
print("RESUMEN DE INDICADORES - COLEGIO LAURA HERRERA")
print("="*70)

print("\n🔵 INDICADORES AMBIENTALES:")
for i, (categoria, meta, actual) in enumerate(zip(categorias_amb, meta_amb, actual_amb)):
    progreso = (actual/meta)*100 if meta > 0 else 0
    barra = "█" * int(progreso/5) + "░" * (20 - int(progreso/5))
    print(f"   {categoria:<20} {actual:>2}% / {meta:>2}% [{barra}] {progreso:.1f}%")

print("\n🔴 INDICADORES SST:")
for i, (categoria, valor) in enumerate(zip(categorias_sst, valores_sst)):
    estado = "✅ CUMPLIDO" if valor >= 100 else "🟡 EN PROCESO" if valor >= 50 else "🔴 PENDIENTE"
    barra = "█" * int(valor/5) + "░" * (20 - int(valor/5))
    print(f"   {categoria:<20} {valor:>3}% [{barra}] {estado}")

print("\n" + "="*70)
print("PERIODO DE EVALUACIÓN: Enero - Diciembre 2025")
print("LOCALIDAD: Bosa - Bogotá D.C.")
print("="*70)
```

```python
# GRÁFICO EVOLUCIÓN TEMPORAL - PROYECCIÓN 12 MESES
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter

# Configuración
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Datos de proyección ambiental (acumulativo)
residuos = [5, 10, 15, 20, 25, 30, 30, 30, 30, 30, 30, 30]
energia = [3, 6, 9, 12, 15, 15, 15, 15, 15, 15, 15, 15]
areas_verdes = [2, 5, 8, 12, 15, 18, 20, 20, 20, 20, 20, 20]

fig, ax = plt.subplots(figsize=(14, 8))

# Gráfico de líneas evolutivas
ax.plot(meses, residuos, marker='o', linewidth=3, markersize=8,
        label='Reducción Residuos (%)', color='#2E8B57')
ax.plot(meses, energia, marker='s', linewidth=3, markersize=8,
        label='Ahorro Energético (%)', color='#FF8C00')
ax.plot(meses, areas_verdes, marker='^', linewidth=3, markersize=8,
        label='Incremento Áreas Verdes (%)', color='#32CD32')

# Líneas de meta
ax.axhline(y=30, color='#2E8B57', linestyle='--', alpha=0.5, label='Meta Residuos (30%)')
ax.axhline(y=15, color='#FF8C00', linestyle='--', alpha=0.5, label='Meta Energía (15%)')
ax.axhline(y=20, color='#32CD32', linestyle='--', alpha=0.5, label='Meta Áreas Verdes (20%)')

# Personalización
ax.set_title('EVOLUCIÓN PROYECTADA - INDICADORES AMBIENTALES 2025\nColegio Laura Herrera de Varela - Bosa',
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Meses del Año')
ax.set_ylabel('Porcentaje de Avance (%)')
ax.legend(loc='lower right')
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 35)

# Añadir anotaciones
ax.annotate('Implementación\nPuntos Ecológicos', xy=(5, 30), xytext=(3, 32),
            arrowprops=dict(arrowstyle='->', color='red'), fontweight='bold')

ax.annotate('Inicio Huertas\nVerticales', xy=(2, 5), xytext=(0, 10),
            arrowprops=dict(arrowstyle='->', color='blue'), fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# GRÁFICO COMPARATIVO ENTRE EDIFICIOS
import pandas as pd

# Datos simulados por edificio
data = {
    'Edificio': ['A - Administrativo', 'B - Primaria', 'C - Bachillerato', 'D - Laboratorios'],
    'Consumo_Energia_kWh': [450, 620, 580, 710],
    'Residuos_kg_semana': [85, 120, 110, 95],
    'Areas_Verdes_m2': [150, 80, 120, 60],
    'Reportes_SST': [2, 5, 3, 8]
}

df = pd.DataFrame(data)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Gráfico 1: Consumo de energía por edificio
colores_edificios = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
barras_energia = ax1.bar(df['Edificio'], df['Consumo_Energia_kWh'], color=colores_edificios)
ax1.set_title('CONSUMO ENERGÉTICO POR EDIFICIO', fontweight='bold')
ax1.set_ylabel('kWh/mes')
ax1.tick_params(axis='x', rotation=15)
for i, v in enumerate(df['Consumo_Energia_kWh']):
    ax1.text(i, v + 10, f'{v} kWh', ha='center', fontweight='bold')

# Gráfico 2: Generación de residuos
barras_residuos = ax2.bar(df['Edificio'], df['Residuos_kg_semana'], color=colores_edificios)
ax2.set_title('GENERACIÓN DE RESIDUOS POR EDIFICIO', fontweight='bold')
ax2.set_ylabel('kg/semana')
ax2.tick_params(axis='x', rotation=15)
for i, v in enumerate(df['Residuos_kg_semana']):
    ax2.text(i, v + 2, f'{v} kg', ha='center', fontweight='bold')

# Gráfico 3: Áreas verdes
barras_areas = ax3.bar(df['Edificio'], df['Areas_Verdes_m2'], color=colores_edificios)
ax3.set_title('ÁREAS VERDES POR EDIFICIO', fontweight='bold')
ax3.set_ylabel('m²')
ax3.tick_params(axis='x', rotation=15)
for i, v in enumerate(df['Areas_Verdes_m2']):
    ax3.text(i, v + 2, f'{v} m²', ha='center', fontweight='bold')

# Gráfico 4: Reportes SST
barras_sst = ax4.bar(df['Edificio'], df['Reportes_SST'], color=colores_edificios)
ax4.set_title('REPORTES SST POR EDIFICIO', fontweight='bold')
ax4.set_ylabel('Número de reportes')
ax4.tick_params(axis='x', rotation=15)
for i, v in enumerate(df['Reportes_SST']):
    ax4.text(i, v + 0.1, f'{v}', ha='center', fontweight='bold')

plt.suptitle('ANÁLISIS COMPARATIVO ENTRE EDIFICIOS - COLEGIO LAURA HERRERA',
             fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

# RESUMEN ESTADÍSTICO
print("📊 RESUMEN ESTADÍSTICO POR INDICADOR:")
print("="*50)
print(f"🔸 Consumo Energético Promedio: {df['Consumo_Energia_kWh'].mean():.1f} kWh")
print(f"🔸 Residuos Generados Promedio: {df['Residuos_kg_semana'].mean():.1f} kg/semana")
print(f"🔸 Áreas Verdes Promedio: {df['Areas_Verdes_m2'].mean():.1f} m²")
print(f"🔸 Reportes SST Totales: {df['Reportes_SST'].sum()}")
print("="*50)
```
